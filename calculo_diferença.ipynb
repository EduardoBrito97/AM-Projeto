```python
import pandas as pd
```

```python
table4 = pd.read_csv ('/content/tabela4.csv')
table4 = table4.set_index('Datasets')
table4
```

```python
table5 = pd.read_csv ('/content/tabela_5_artigo.csv')
table5 = table5.set_index('Datasets')
table5 = table5.drop('GP-kSS', 1)
table5
```

```python
table_article = pd.concat([table5,table4],axis=1)
table_article = table_article.drop('WAF(CC)', 1)
table_article = table_article.drop('WAF(CD)', 1)
table_article = table_article.drop('GFRNN', 1)
table_article = table_article.drop('Im-GFRNN', 1)
table_article

```

```python
our = pd.read_csv ('/content/all_algorithms.csv')
our = our.set_index('Dataset')
our
```

```python
compare = table_article - our
compare = compare.round(4)
compare
```

```python
print(compare.to_csv(index=True))
```
